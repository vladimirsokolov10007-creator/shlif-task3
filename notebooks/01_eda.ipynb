# EDA — шлифы (task-3)
Распределение классов, размеры изображений, примеры. Требует `data/processed/index.csv` (см. `python -m src.data.index`).

```python
import sys, os
sys.path.append(os.path.abspath('..'))
import pandas as pd, numpy as np
import matplotlib.pyplot as plt
from PIL import Image
Image.MAX_IMAGE_PIXELS = None
from src.utils import load_config, resolve
cfg = load_config('../configs/config.yaml')
df = pd.read_csv(resolve(cfg['paths']['index_csv']))
df.head()
```

```python
photos = df[(df.part=='photo') & (df.class_id>=0)]
counts = photos['class'].value_counts()
print(counts)
counts.plot(kind='bar', title='Фото по классам'); plt.tight_layout(); plt.show()
```

```python
sized = photos[photos.width>0]
plt.scatter(sized.width, sized.height, s=6, alpha=0.4)
plt.xlabel('width'); plt.ylabel('height'); plt.title('Размеры фото'); plt.show()
print('панорамы:'); print(df[df.part=='panorama'][['rel','width','height']])
```

```python
# Примеры по одному на класс
fig, axes = plt.subplots(1, len(cfg['data']['classes']), figsize=(16,4))
for ax, cls in zip(np.atleast_1d(axes), cfg['data']['classes']):
    sub = photos[photos['class']==cls]
    if len(sub)==0: ax.axis('off'); continue
    p = sub.sample(1, random_state=0).iloc[0]['path']
    ax.imshow(Image.open(p).convert('RGB')); ax.set_title(cls); ax.axis('off')
plt.tight_layout(); plt.show()
```
